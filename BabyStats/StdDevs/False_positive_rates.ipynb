```python
%matplotlib inline

from pylab import *

#from matplotlib import pyplot as plt
import matplotlib.pyplot  as plt
import matplotlib.ticker
#from matplotlib.ticker import ScalarFormatter

import sys, platform, os
import numpy as np
import math
from scipy.special import erf
from scipy.special import erfc
```

```python
#sigs = list(range(11, 17))
sigs_in = np.arange(0, 10, 0.1)
#print(sigs_in)
print(type(sigs_in))
```

```python
m = (sigs_in/math.sqrt(2.))
type(sigs_in), type(m)

#sampleSize = (1./ (erfc(m)))
sampleSize = (1./ (erfc(sigs_in/math.sqrt(2))))
samplePower = sampleSize**(2.37)
log_sampleSize = np.log10(sampleSize)
#print(sigs_in, sampleSize, np.log10(sampleSize))
```

```python
matplotlib.rcParams.update({'font.size': 12})

plt.figure()
plt.subplots(1, 1, figsize=(9,6))
#plt.title('False Positive Rate (FPR)', size='x-large')
#fig, ax = plt.subplots()

## 

ax = subplot(111)
ax.set_xscale('log')
#ax.set_yscale('log')

## Setting up the axes and all that...
plt.xlabel('Sample Size', size='xx-large')
#plt.xlabel('log(Sample Size)', size='x-large')
#plt.axis([1, 10000000])

plt.xlim([1.,1e10])
plt.xscale('log')

#plt.set_xticks([0, 100, 10000, ])
#for axis in plt.xlim:
 #   plt.set_major_formatter(ScalarFormatter())
#plt.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())

## http://stackoverflow.com/questions/6431248/matplotlib-logarithmic-scale-but-require-non-logarithmic-labels
#ax.xaxis.set_major_formatter(matplotlib.ticker.ScalarFormatter())
#ax.xaxis.set_major_formatter(matplotlib.ticker.FormatStrFormatter('%d'))
#ax.yaxis.set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, pos: str(int(round(x)))))

plt.tick_params(labelsize=6)
x = np.array([1,1000,1e6,1e9])
#y = np.array([20,21,22,23])
my_xticks = ['1','1,000','1 million','1 billion']
plt.xticks(x, my_xticks)

#plt.ylabel('Sigmas, $\sigma$', size='xx-large')
plt.ylabel('Time to discovery (in minutes)', size='x-large')
plt.ylim([0.0,8.5])


plt.grid(True, 'major', 'y', ls='--', lw=.5, c='k', alpha=.3)
##
plt.plot(sampleSize, sigs_in, linewidth=5, label='Traditional', color='red')
#plt.plot(log_sampleSize, sigs_in, linewidth=2, label='False Positive rate', color='red')
plt.plot(samplePower, sigs_in, linewidth=5, label='Machine Learning', color='blue', ls='dashed')

plt.tick_params(axis='both', which='major', labelsize=14)
plt.legend(loc=(0.05,0.75), scatterpoints=1, fontsize=16)

plt.savefig('False_positive_rates_temp.jpeg', bbox_inches='tight', dpi=800)
```
